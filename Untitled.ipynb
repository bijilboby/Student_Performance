```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

# 1. Load and explore the data
df = pd.read_csv('StudentsPerformance.csv')
print("First 5 rows:")
print(df.head())
```

```python
# 2. Create target variable (average score)
def score_to_grade(avg_score):
    if avg_score >= 90: return 'A (Excellent)'
    elif avg_score >= 80: return 'B (Very Good)'
    elif avg_score >= 70: return 'C (Good)'
    elif avg_score >= 60: return 'D (Needs Improvement)'
    else: return 'F (Work Hard)'
df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
df['grade'] = df['average_score'].apply(score_to_grade)
```

```python
# 3. Select features and target
features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
X = df[features]  # Features
y = df['average_score']  # Target
```

```python
# 4. Convert categorical data to numbers
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)
```

```python
# 5. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# 6. Create and train the model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
# 7. Make predictions and evaluate
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print(f"\nModel Evaluation:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Average Error: {mse**0.5:.2f} points")
```

```python
# 8. Show feature importance
feature_names = encoder.get_feature_names_out(features)
importances = model.feature_importances_

plt.figure(figsize=(10, 5))
plt.barh(feature_names, importances)
plt.title("Feature Importance")
plt.show()
```

```python
# 9. Make a sample prediction
sample_student = pd.DataFrame({
    'gender': ['female'],
    'race/ethnicity': ['group B'],
    'parental level of education': ["bachelor's degree"],
    'lunch': ['standard'],
    'test preparation course': ['completed']
})

sample_encoded = encoder.transform(sample_student)
predicted_score = model.predict(sample_encoded)[0]
print(f"\nSample Prediction:")
print(f"Predicted Average Score: {predicted_score:.1f}/100")
gradepred = score_to_grade(predicted_score)
print(f"The grade of student is {gradepred}")
```
